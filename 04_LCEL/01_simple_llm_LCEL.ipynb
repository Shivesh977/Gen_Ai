### Build a Simple LLM Application with LCEL
This application will translate text from English into another language.It's just a single LLM call plus some prompting

```python
import os 
from dotenv import load_dotenv 
load_dotenv

groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
!pip install langchain_groq
```

```python
# from langchain_openai import ChatOpenAi # for openai key 
from langchain_groq import ChatGroq

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=groq_api_key
)
```

```python
!pip install langchain_core
```

```python
from langchain_core.messages import HumanMessage,SystemMessage
# Humanmessage is : input given to llm model 
# System message is : instruction given to llm model on how to behave ..
# To differentiate between them we use Humanmessage and systemmessage 

messages =[ # how llm model should behave 
           SystemMessage(content="Translate the following from english to french "),
           HumanMessage(content="Hello how are you") # this sentence need to be converted into french 
]

result=model.invoke(messages)
```

```python
# to retrieve only output 
from langchain_core.output_parsers  import StrOutputParser 

parser=StrOutputParser()

parser.invoke(result)
```

```python
# Using LCEL - we can chain the components

chain=model|parser   # chaining ....
chain.invoke(messages)
```

```python
# Prompt templetas
# generalizing .....instead of converting into french user can convert into various language based on its mood 
# so we are making a generic  prompt template 

from langchain_core.prompts import ChatPromptTemplate

generic_template="Translate the following into {language}"

prompt=ChatPromptTemplate.from_messages(
    [("system",generic_template),("user","{text}")]
)

```

```python
result=prompt.invoke({"language" : "French ","text":"Hello"})
```

```python
chain=prompt|model|parser # chaining using LCEL 
chain.invoke({"language":"French","text":"Hello"})
```
